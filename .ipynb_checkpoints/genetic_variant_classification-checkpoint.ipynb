```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import boxcox
from sklearn.preprocessing import MinMaxScaler
```

# Genetic Variant Classification

## Loading the data

```python
data = pd.read_csv('data/clinvar_conflicting.csv')
```

```python
data.head()
```

## Cleaning the data

There are columns that can be dropped as they are not related to the problem at hand

```python
data = data.drop(columns = ['BAM_EDIT', 'EXON', 'CLNDISDB', 'CLNHGVS', 
         'MC', 'CLNVI', 'SYMBOL', 'Feature', 'Feature_type', 
         'BIOTYPE', 'INTRON', 'CADD_RAW'])
```

#### Exploring nan values

```python
missing_perc = pd.DataFrame(data.isna().mean(), columns=['missing_percentage'])
missing_perc=missing_perc.sort_values(by='missing_percentage', ascending=False)
missing_perc
```

We can drop the columns with over 99 percent missing data

```python
to_drop = missing_perc.index[:9]
to_drop
```

```python
data = data.drop(columns=to_drop)
```

```python
data
```

```python
data.info()
```

As the order of the rows is determined by first the chromosome that the variant is on and then the position of the allele, we will fill the missing values in the below rows with forward fill, as the best logical guess we can make is that the closer the position, the more related the values.

For the other numerical values, we can interpolate.

```python
data.select_dtypes(include=['object'])
```

data['CHROM'] must be cleaned as there are ints and strings mixed

```python
data['CHROM'].unique()
```

```python
chrom_map = {2: '2', 3: '3', 11: '11', 12: '12',13: '13',14: '14', 15: '15',16: '16', 17: '17', 18: '18', 19: '19', 
             20: '20', 21: '21', 22: '22'}
```

```python
data['CHROM'] = data['CHROM'].replace(chrom_map)
```

```python
data['CHROM'].unique()
```

Setting some 'unknown' values to nan before imputing

```python
data.loc[data[data['PolyPhen']=='unknown'].index, 'PolyPhen'] = np.nan
```

```python
for column in data.columns:
    if column == 'BLOSUM62':
        data[column] = data[column].fillna(method='ffill')
    elif data[column].dtype == 'object':
        data[column] = data[column].fillna(method='ffill')
    elif data[column].dtype == 'float64':
        data[column] = data[column].interpolate()

data['LoFtool'] = data['LoFtool'].fillna(0)

data.info()
```

```python
data.select_dtypes(include=['object'])
```

Here we are mapping some encoding some key features that do have a relationship with each other

```python
SIFT_map = {'tolerated': 1, 'tolerated_low_confidence': 2, 'deleterious_low_confidence': 3, 'deleterious': 4}
PolyPhen_map = {'benign': 1, 'possibly_damaging': 2, 'probably_damaging': 3}
IMPACT_map = {'LOW': 1, "MODERATE": 2, "HIGH": 3, "MODIFIER": 4}
```

```python
data['SIFT'] = data['SIFT'].replace(SIFT_map)
data['PolyPhen'] = data['PolyPhen'].replace(PolyPhen_map)
data['IMPACT'] = data['IMPACT'].replace(IMPACT_map)
```

Remaining object features can be label encoded. It is not ideal as they may not have an inherent order to their relationship, but with limited resources, if we created as many columns as necessary to one hot encode the remaining features it would be too computationally expensive. 

```python
label_encoder = LabelEncoder()

for column in data.columns:
    if data[column].dtype == 'object':
        data[column] = label_encoder.fit_transform(data[column])

data.info()
```

```python
num_columns = 3
num_rows = (len(data.columns) - 1) // num_columns + 1

fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, 40))

if num_rows == 1:
    axes = axes.reshape(1, -1)    

for i, column in enumerate(data.columns):
    row = i // num_columns
    col = i % num_columns
    ax = axes[row][col]
    sns.boxplot(x = data[column], ax=ax)
    ax.set_title(column)
    
for i in range(len(data.columns), num_rows * num_columns):
    row = i // num_columns
    col = i % num_columns
    axes[row][col].axis('off')

plt.tight_layout()
plt.show()
```

As shown here, the columns AF_ESP, AF_EXAC, and AF_TGP appear to be comprised mostly of outliers, which is concerning. Upon further reading, it seems that missing values in these columns were assigned a '0' instead of Nan, so we must remedy that before attempting any analysis. We will interpolate for the ) value rows.

```python
allele_data = data[['AF_ESP', 'AF_EXAC', 'AF_TGP']]

for column in allele_data.columns:
    allele_data.loc[allele_data[column] == 0, column] = None
    
allele_data = allele_data.interpolate()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for i, column in enumerate(allele_data.columns):
    sns.boxplot(x = allele_data[column], ax=axes[i])
    ax.set_title(column)

plt.tight_layout()
plt.show()
```

Even after we interpolate, further transformation of the outliers is necessary. We will use the Box-Cox transformation as it is the most appropriate for the data. 

```python
allele_data_boxcox = allele_data
for column in allele_data_boxcox.columns:
    transformed_data, _ = boxcox(allele_data_boxcox[column].dropna() + 1)
    allele_data_boxcox[column].loc[~allele_data_boxcox[column].isna()] = transformed_data
```

```python
allele_data_boxcox
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for i, column in enumerate(allele_data_boxcox.columns):
    sns.boxplot(x = allele_data_boxcox[column], ax=axes[i])
    ax.set_title(column)

plt.tight_layout()
plt.show()
```

Transformation looks decent, so we can replace our full DataFrame with these transformed values

```python
data.update(allele_data_boxcox)
data.head()
```

```python
num_columns = 3
num_rows = (len(data.columns) - 1) // num_columns + 1

fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, 40))

if num_rows == 1:
    axes = axes.reshape(1, -1)    

for i, column in enumerate(data.columns):
    row = i // num_columns
    col = i % num_columns
    ax = axes[row][col]
    sns.boxplot(x = data[column], ax=ax)
    ax.set_title(column)

for i in range(len(data.columns), num_rows * num_columns):
    row = i // num_columns
    col = i % num_columns
    axes[row][col].axis('off')
    
plt.tight_layout()
plt.show()
```

There are some further outliers, but the plots look good enough to proceed at this point.

```python
num_columns = 3
num_rows = (len(data.columns) - 1) // num_columns + 1

fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, 40))

if num_rows == 1:
    axes = axes.reshape(1, -1)

for i, column in enumerate(data.columns):
    row = i // num_columns
    col = i % num_columns
    ax = axes[row][col]
    sns.histplot(data[column], kde=True, bins=20, color='skyblue', ax=ax)
    ax.set_title(column)

for i in range(len(data.columns), num_rows * num_columns):
    row = i // num_columns
    col = i % num_columns
    axes[row][col].axis('off')

plt.tight_layout()
plt.show()
```
